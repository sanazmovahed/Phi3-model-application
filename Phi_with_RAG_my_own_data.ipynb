```python
# prompt: load a phi3 model from huggingface

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load Phi-3 model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3.5-mini-instruct")
model = AutoModelForCausalLM.from_pretrained("microsoft/Phi-3.5-mini-instruct", torch_dtype=torch.float16, device_map="auto")
```

```python
from transformers import pipeline

# Create a text generation pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Generate text
response = pipe("Tell me a joke!", max_new_tokens=100)
print(response[0]["generated_text"])
```

```python
import pypdf
pdf_path = "/content/sanaz-movahed-academic-cv-updated-January-2025.pdf"
def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF file."""
    pdf_reader = pypdf.PdfReader(pdf_path)
    text = ""
    for page in pdf_reader.pages:
        text += page.extract_text() + "\n"
    return text

# Example usage
pdf_text = extract_text_from_pdf(pdf_path)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load embedding model
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Split text into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_text(pdf_text)

# Store chunks in FAISS
vectorstore = FAISS.from_texts(docs, embedding_model)
```

```python
from transformers import pipeline

# Load Phi-3 model
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

def query_pdf(question):
    """Finds relevant context and queries Phi-3."""
    similar_docs = vectorstore.similarity_search(question, k=3)  # Get top 3 relevant chunks
    context = "\n".join([doc.page_content for doc in similar_docs])

    # Form the prompt
    prompt = f"Context:\n{context}\n\nQuestion: {question}\nAnswer:"

    # Generate response
    response = pipe(prompt, max_new_tokens=500)
    return response[0]["generated_text"]

# Example query
print(query_pdf("What is the main topic of the document?"))
```
